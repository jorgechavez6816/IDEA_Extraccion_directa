```python
import win32com.client as win32ComClient

if __name__ == "__main__":
    try:
        dbName = "Ejemplo-Detalle de ventas.IMD"
        resultDbName = "Extraccion_Cuarentena.IMD"
        criteria = ' (TOTAL  >= 300 .AND.  TOTAL  <= 10500) .AND. (COD_PROD  = "04" .OR. COD_PROD  = "05" .OR. COD_PROD  = "06") .AND. (FECHA_FACT  >= "20150316" .AND. FECHA_FACT  <= "20150624")'
        idea = win32ComClient.Dispatch(dispatch="Idea.IdeaClient")
        db = idea.OpenDatabase(dbName)
        task = db.Extraction()
        task.IncludeAllFields()
        task.AddExtraction(resultDbName, "", criteria)
        task.PerformTask(1, db.Count)
        idea.OpenDatabase(resultDbName)
    finally:
        task = None
        db = None
        idea = None
        
```
